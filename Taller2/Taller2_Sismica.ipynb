#Arreglo 1

```python
#Se extraen las coordenadas
A1_data_Coords=np.loadtxt('Taller2_Coordenadas_A1.csv', delimiter=';', skiprows=1)
#data=[Station, East, North, Elevation]
#Se calcula un arreglo de distancias
A1_distance=A1_data_Coords[:, 1]-A1_data_Coords[29, 1] 
#Se extrae el arreglo de tiempos
A1_data_Picks=np.loadtxt('Array1_SP01_fb.txt', skiprows=7)
#data=[#, X, T, Layer]
#Se toman los tiempos no nulos
A1_times=A1_data_Picks[:, 2][np.where(A1_data_Picks[:, 2]>=0.)]
#Se toman las distancias negativas correspondientes a los tiempos no nulos.
A1_distance=A1_distance[np.where(A1_data_Picks[:, 2]>=0.)and A1_distance<=0.]
A1_times=A1_times[:len(A1_distance)]
#Se grafica
plt.scatter(A1_distance, A1_times)
plt.gca().invert_yaxis() 
#plt.grid()
plt.xlabel('$Offset (m)$', fontsize=15)
plt.ylabel('$t (s)$', fontsize=15)
#plt.gcf()
#plt.savefig('picadosArreglo1.png', bbox_inches='tight')
```

```python
#La observación de la gráfica nos permite ver que el cambio en la pendiente se da alrededor de los -14m o los -22m
A1_xc=A1_distance[-12] #cambiar el índice por -8 para ver el resultado cuando x_C=-14m
print A1_xc
x=np.linspace(-60, 0, 100)
#Se hacen las regresiones lineales para ambas rectas.
A1_DW_dist=A1_distance[np.where(A1_distance>=A1_xc)]
A1_DW_time=A1_times[np.where(A1_distance>=A1_xc)]
A1_DW_slope, A1_DW_intercept, A1_DW_r_value, p_value, std_err = linregress(A1_DW_dist,A1_DW_time)

A1_RW_dist=A1_distance[np.where(A1_distance<=A1_xc)]
A1_RW_time=A1_times[np.where(A1_distance<=A1_xc)]
A1_RW_slope, A1_RW_intercept, A1_RW_r_value, p_value, std_err = linregress(A1_RW_dist,A1_RW_time)
#Se calculan las velocidades vw, vb
A1_vw=A1_DW_slope**-1
A1_vb=A1_RW_slope**-1
#Se calcula la profundidad de la capa meteorizada
A1_zw=0.5*np.sqrt((A1_vb-A1_vw)/(A1_vb+A1_vw))*A1_xc

#Se grafica con el eje de tiempo invertido
A1_slope_change, ax = plt.subplots(1, 1, figsize=(7, 5))
ax.invert_yaxis()
ax.scatter(A1_DW_dist, A1_DW_time, c='r', s=50)
ax.plot(x, x*A1_DW_slope+A1_DW_intercept, c='r', label='$v_w=%d m/s$, $r^2=%.3f$'%(-A1_DW_slope**-1,A1_DW_r_value**2))
ax.scatter(A1_RW_dist, A1_RW_time, c='b', s=50)
ax.plot(x, x*A1_RW_slope+A1_RW_intercept, c='b', label='$v_b=%d m/s$, $r^2=%.3f$'%(-A1_RW_slope**-1,A1_RW_r_value**2))
ax.set_xlabel('$Offset (m)$', fontsize=15)
ax.set_ylabel('$t (s)$', fontsize=15)
plt.legend(fontsize=15, loc=0)
#A1_slope_change.savefig('rectasVelocidadesArr1.png', bbox_inches='tight')
#Se imprimen algunos valores que incluyen los coeficientes de correlación para ambas rectas.
print 'DW r^2= ', A1_DW_r_value**2, '\n', 'RW r^2= ', A1_RW_r_value**2, '\n', 'ZW= ', A1_zw
```

```python
#Como se obtuvo un vw distinto en ambos 'shots' se tomará como vw un promedio de los dos:
A2_vw=np.mean([A2_vw_1, A2_vw_2])

#Se calcula el ángulo phi de buzamiento de la capa.
def calculate_phi(vw, vbplus, vbminus):
    return 0.5*(np.arcsin(vw/vbminus)-np.arcsin(vw/vbplus))
A2_phi=calculate_phi(A2_vw, A2_vb_2, A2_vb_1)

print 'El ángulo de buzamiento es phi= ', np.rad2deg(A2_phi), 'grados.'

#Se calcula la velocidad vb real

def calculate_real_vb(phi, vbplus, vbminus):
    return 2.*np.cos(phi)/(vbplus**-1+vbminus**-1)

A2_vb=calculate_real_vb(A2_phi, A2_vb_2, A2_vb_1)
print 'La velocidad real de la capa base es vb= ', A2_vb, 'm/s.'

#Se calcula el espesor de la capa meteorizada

def calculate_w_depth(vb, vw, ti, phi):
    return vb*vw*ti/(2.*np.cos(phi)*np.sqrt(vb**2-vw**2))

A2_zw_src=calculate_w_depth(A2_vb, A2_vw, A2_RW_intercept_1, A2_phi)
A2_zw_rcv=calculate_w_depth(A2_vb, A2_vw, A2_RW_intercept_2, A2_phi)
print 'El espesor de la capa meteorizada bajo el primer sensor es Zw1= ', A2_zw_src, 'm. \n' \
'El espesor de la capa meteorizada bajo el último sensor es Zw2= ', A2_zw_rcv, 'm.'
```
